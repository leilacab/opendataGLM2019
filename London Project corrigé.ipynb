```python
import csv
```

```python
points_in_zone = csv.reader (open ('/Users/Leila/Desktop/ComputationalMethods/LondonData/pointsinzone.csv','r',encoding='utf8'),delimiter=',')
headers = points_in_zone.__next__() 
zone_list=[row[0] for row in points_in_zone]
```

```python
all_points = csv.reader (open ('/Users/Leila/Desktop/ComputationalMethods/LondonData/dft_aadf_region_id_6.csv',\
                              'r',encoding='utf8'),delimiter=',')
all_points_headers = all_points.__next__()

bd_zone= {}
geo= {}
no_zone_list=[]
for row in all_points:
    row_helper = { all_points_headers[i] : row[i] for i in range(len(all_points_headers)) }
    CPI,year,nvehicle=row_helper ['count_point_id'],row_helper ['year'],row_helper ['all_motor_vehicles']
    lat,long=row_helper ['latitude'],row_helper ['longitude']
    if CPI not in bd_zone : bd_zone[CPI]= {} 
    bd_zone[CPI][year]=nvehicle
    if CPI not in geo : geo[CPI]= {} 
    geo[CPI]['lat']=float(lat)
    geo[CPI]['long']=float(long)
    geo[CPI]['zone']=CPI in zone_list #condition vrai faux 
    #now creating the list with all the points outside the zone
    if CPI not in zone_list : no_zone_list.append(CPI)
```

```python
##Now calculate distances

# Prof code part to set up the function
import numpy as np
import math

    # and a quick converter
def deg2rad(deg): return (deg*(np.pi/180))

    # and here you are !
def latlon2km (lat1,lon1,lat2,lon2):
    R = 6371
    dLat = deg2rad(lat2-lat1)
    dLon = deg2rad(lon2-lon1) 
    a = np.sin(dLat/2)*np.sin(dLat/2)+np.cos(deg2rad(lat1))*np.cos(deg2rad(lat2))*np.sin(dLon/2)*np.sin(dLon/2)
    c = 2 * math.atan2(np.sqrt(a),np.sqrt(1-a)) 
    return(R*c)
```

```python
# First loop : for all points in the zone we calculate the distance to all the points outside the zone and keep in storage only the closest point ID and distance
for CPI in zone_list:
    mini_CPI=-1
    min_dist_CPI=1000000
    for CPI2 in no_zone_list: 
        new_dist=latlon2km(geo[CPI]['lat'],geo[CPI]['long'],geo[CPI2]['lat'],geo[CPI2]['long'])
        if new_dist < min_dist_CPI : 
            min_dist_CPI,mini_CPI=new_dist,CPI2
            geo[CPI]['min_dist_CPI']=min_dist_CPI
            geo[CPI]['mini_CPI']=mini_CPI
```

```python
# Second loop : for all points in the outside the zone we calculate the distance to all the points in the zone and keep in storage only the closest point ID and distance
for CPI in no_zone_list:
    mini_CPI=-1
    min_dist_CPI=1000000
    for CPI2 in zone_list: 
        new_dist=latlon2km(geo[CPI]['lat'],geo[CPI]['long'],geo[CPI2]['lat'],geo[CPI2]['long'])
        if new_dist < min_dist_CPI : 
            min_dist_CPI,mini_CPI=new_dist,CPI2
            geo[CPI]['min_dist_CPI']=min_dist_CPI
            geo[CPI]['mini_CPI']=mini_CPI
```
